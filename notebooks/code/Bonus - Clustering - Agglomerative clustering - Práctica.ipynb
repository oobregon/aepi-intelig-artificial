```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
```

```python
file_datapath = "C:/Users/user/Desktop/water-treatment.data"
df = pd.read_csv(file_datapath)
df
```

```python
df.set_index(df["D-1/3/90"], drop=True, inplace=True)
```

```python
df.drop("D-1/3/90", axis=1, inplace = True)
```

```python
df
```

```python
df.dtypes
```

```python
df.replace("?", "", inplace=True)
```

```python
df
```

```python
df = df.apply(pd.to_numeric)
```

```python
df
```

```python
df.mean()
```

```python
df = df.fillna(df.mean())
df
```

```python
df.dtypes
```

```python
agc = AgglomerativeClustering(n_clusters=13, 
                              affinity='euclidean', 
                              linkage='ward', 
                              distance_threshold=None)
```

```python
agc.fit(df)
```

```python
agc.labels_
```

```python
df["cluster_id"] = agc.labels_
df
```
